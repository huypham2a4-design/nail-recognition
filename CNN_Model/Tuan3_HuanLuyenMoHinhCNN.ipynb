```python
# ============================================================
#  Mô hình CNN tự build với TensorFlow / Keras
# ============================================================

# ── STEP 0: Import Libraries ─────────────────────────────────
import os
import warnings
warnings.filterwarnings("ignore")

import cv2
from tqdm import tqdm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import shuffle

print("TensorFlow version:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))


# ── STEP 1: Cấu hình ─────────────────────────────────────────
IMAGE_SIZE   = (150, 150)
BATCH_SIZE   = 32
EPOCHS       = 50
NUM_CLASSES  = 3
RANDOM_SEED  = 42

CLASS_NAMES  = ['Long_Nail', 'Painted_Nail', 'Short_Nail']
CLASS_LABELS = {c: i for i, c in enumerate(CLASS_NAMES)}

TRAIN_DIR = '/content/drive/MyDrive/DataDeeplearning/DataSet/Train'
VAL_DIR   = '/content/drive/MyDrive/DataDeeplearning/DataSet/Val'
MODEL_SAVE_PATH = 'nail_cnn_model.h5'

print("\nClass mapping:", CLASS_LABELS)


# ── STEP 2: Load & Preprocess Dataset ────────────────────────
def load_dataset(dataset_dir, class_names, image_size=(150, 150), split_name="Dataset"):
    """
    Đọc ảnh từ thư mục con theo từng class.
    Trả về (Images: float32 array, Labels: int32 array)
    """
    Images, Labels = [], []

    for class_name in class_names:
        class_path = os.path.join(dataset_dir, class_name)
        if not os.path.exists(class_path):
            print(f"  [WARNING] Không tìm thấy thư mục: {class_path}")
            continue

        files = [f for f in os.listdir(class_path)
                 if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp'))]

        print(f"  [{split_name}] {class_name}: {len(files)} ảnh")

        for file in tqdm(files, desc=f"  Loading {class_name}", leave=False):
            img_path = os.path.join(class_path, file)
            img = cv2.imread(img_path)
            if img is None:
                continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # BGR → RGB
            img = cv2.resize(img, image_size)
            Images.append(img)
            Labels.append(CLASS_LABELS[class_name])

    Images = np.array(Images, dtype='float32')
    Labels = np.array(Labels, dtype='int32')
    return Images, Labels


print("\n Đang load dữ liệu...")
train_images, train_labels = load_dataset(TRAIN_DIR, CLASS_NAMES, IMAGE_SIZE, "Train")
val_images,   val_labels   = load_dataset(VAL_DIR,   CLASS_NAMES, IMAGE_SIZE, "Val")

# Normalize [0,255] → [0.0, 1.0]
train_images = train_images / 255.0
val_images   = val_images   / 255.0

# Shuffle training set
train_images, train_labels = shuffle(train_images, train_labels, random_state=RANDOM_SEED)

print(f"\n Train: {train_images.shape}  |  Labels: {train_labels.shape}")
print(f" Val  : {val_images.shape}    |  Labels: {val_labels.shape}")


# ── STEP 3: EDA (Exploratory Data Analysis) ──────────────────
def plot_eda(train_labels, val_labels, class_names, train_images):
    """Vẽ phân phối class và mẫu ảnh."""
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("EDA — Phân tích dữ liệu móng tay", fontsize=16, fontweight='bold')

    # 1. Bar chart — train distribution
    unique, counts = np.unique(train_labels, return_counts=True)
    axes[0].bar([class_names[i] for i in unique], counts,
                color=['#E74C3C', '#3498DB', '#2ECC71'], edgecolor='black')
    axes[0].set_title("Phân phối lớp — Train")
    axes[0].set_ylabel("Số lượng ảnh")
    for i, v in enumerate(counts):
        axes[0].text(i, v + 5, str(v), ha='center', fontweight='bold')

    # 2. Bar chart — val distribution
    unique_v, counts_v = np.unique(val_labels, return_counts=True)
    axes[1].bar([class_names[i] for i in unique_v], counts_v,
                color=['#E74C3C', '#3498DB', '#2ECC71'], edgecolor='black', alpha=0.8)
    axes[1].set_title("Phân phối lớp — Val")
    axes[1].set_ylabel("Số lượng ảnh")
    for i, v in enumerate(counts_v):
        axes[1].text(i, v + 2, str(v), ha='center', fontweight='bold')

    # 3. Pie chart — train
    axes[2].pie(counts, labels=[class_names[i] for i in unique],
                autopct='%1.1f%%', colors=['#E74C3C', '#3498DB', '#2ECC71'],
                startangle=90)
    axes[2].set_title("Tỉ lệ lớp — Train")

    plt.tight_layout()
    plt.savefig('eda_distribution.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("  [Saved] eda_distribution.png")

    # Sample images
    fig2, axes2 = plt.subplots(3, 5, figsize=(15, 9))
    fig2.suptitle("Mẫu ảnh từng class", fontsize=14, fontweight='bold')
    for cls_idx, cls_name in enumerate(class_names):
        cls_imgs = train_images[train_labels == cls_idx]
        for col in range(5):
            ax = axes2[cls_idx][col]
            if col < len(cls_imgs):
                ax.imshow(cls_imgs[col])
            ax.axis('off')
            if col == 0:
                ax.set_ylabel(cls_name, fontsize=11, rotation=0,
                              labelpad=80, va='center')
    plt.tight_layout()
    plt.savefig('eda_samples.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("  [Saved] eda_samples.png")


print("\n EDA...")
plot_eda(train_labels, val_labels, CLASS_NAMES, train_images)


# ── STEP 4: Xây dựng mô hình CNN ─────────────────────────────
def build_cnn_model(input_shape=(150, 150, 3), num_classes=3):
    """
    CNN tự build:
      3 khối Conv2D + MaxPooling2D
      → Flatten → Dense → Dropout → Output
    """
    model = Sequential([
        # ── Khối 1: học đặc trưng thấp (cạnh, màu, gradient) ──
        layers.Conv2D(32, (3, 3), activation='relu', padding='same',
                      input_shape=input_shape, name='conv2d_1'),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same',
                      name='conv2d_1b'),
        layers.MaxPooling2D((2, 2), name='pool_1'),       # 150→75

        # ── Khối 2: học đặc trưng trung bình (texture, góc cạnh) ──
        layers.Conv2D(64, (3, 3), activation='relu', padding='same',
                      name='conv2d_2'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same',
                      name='conv2d_2b'),
        layers.MaxPooling2D((2, 2), name='pool_2'),       # 75→37

        # ── Khối 3: học đặc trưng cao (hình dạng, cấu trúc) ──
        layers.Conv2D(128, (3, 3), activation='relu', padding='same',
                      name='conv2d_3'),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same',
                      name='conv2d_3b'),
        layers.MaxPooling2D((2, 2), name='pool_3'),       # 37→18

        # ── Flatten + Fully Connected ──
        layers.Flatten(name='flatten'),

        layers.Dense(512, activation='relu', name='dense_1'),
        layers.Dropout(0.5, name='dropout_1'),

        layers.Dense(256, activation='relu', name='dense_2'),
        layers.Dropout(0.3, name='dropout_2'),

        # ── Output: softmax → 3 lớp ──
        layers.Dense(num_classes, activation='softmax', name='output'),
    ], name='NailCNN')

    return model


model = build_cnn_model(input_shape=(*IMAGE_SIZE, 3), num_classes=NUM_CLASSES)
model.summary()

# Vẽ sơ đồ mô hình (nếu có graphviz)
try:
    keras.utils.plot_model(model, to_file='model_architecture.png',
                           show_shapes=True, show_layer_names=True,
                           expand_nested=True, dpi=96)
    print("  [Saved] model_architecture.png")
except Exception as e:
    print(f"  [Skip] plot_model: {e}")


# ── STEP 5: Compile & Data Augmentation ──────────────────────
# Data Augmentation chỉ áp dụng lên Train
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomBrightness(0.1),
    layers.RandomContrast(0.1),
], name='augmentation')

# Compile
model.compile(
    optimizer=keras.optimizers.Adamax(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_accuracy', patience=10,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                      patience=5, min_lr=1e-6, verbose=1),
    ModelCheckpoint(MODEL_SAVE_PATH, monitor='val_accuracy',
                    save_best_only=True, verbose=1),
]

print("\n  Compile xong. Bắt đầu training...")


# ── STEP 6: Huấn luyện mô hình ───────────────────────────────
# Tạo tf.data pipeline với augmentation
def make_dataset(images, labels, augment=False, batch_size=32):
    ds = tf.data.Dataset.from_tensor_slices((images, labels))
    if augment:
        ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                    num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = make_dataset(train_images, train_labels, augment=True,  batch_size=BATCH_SIZE)
val_ds   = make_dataset(val_images,   val_labels,   augment=False, batch_size=BATCH_SIZE)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)


# ── STEP 7: Vẽ Accuracy & Loss ───────────────────────────────
def plot_accuracy_loss(history):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle("Kết quả huấn luyện mô hình CNN — Nhận diện móng tay",
                 fontsize=14, fontweight='bold')

    # Accuracy
    axes[0].plot(history.history['accuracy'],     '-',  label='Train Accuracy', color='#E67E22', linewidth=2)
    axes[0].plot(history.history['val_accuracy'], '--', label='Val Accuracy',   color='#3498DB', linewidth=2)
    axes[0].set_title("Train vs Val Accuracy")
    axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Accuracy")
    axes[0].legend(); axes[0].grid(True, alpha=0.3)
    axes[0].set_ylim([0, 1.05])

    # Loss
    axes[1].plot(history.history['loss'],     '-',  label='Train Loss', color='#E74C3C', linewidth=2)
    axes[1].plot(history.history['val_loss'], '--', label='Val Loss',   color='#9B59B6', linewidth=2)
    axes[1].set_title("Train vs Val Loss")
    axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Loss")
    axes[1].legend(); axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('training_curves.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("  [Saved] training_curves.png")


plot_accuracy_loss(history)

# Đánh giá trên tập Val
val_loss, val_acc = model.evaluate(val_ds, verbose=0)
print(f"\n Val Loss    : {val_loss:.4f}")
print(f" Val Accuracy: {val_acc:.4f} ({val_acc*100:.2f}%)")


# ── STEP 8: Confusion Matrix & Classification Report ─────────
print("\n Đang tính Confusion Matrix...")

# Dự đoán
predictions  = model.predict(val_images, verbose=0)
pred_labels  = np.argmax(predictions, axis=1)

# Confusion Matrix
CM = confusion_matrix(val_labels, pred_labels)
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(CM, annot=True, fmt='d', cmap='Blues',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES, ax=ax,
            linewidths=0.5, linecolor='gray')
ax.set_title('Confusion Matrix — Nhận diện móng tay', fontsize=14, pad=15)
ax.set_xlabel('Predicted Label', fontsize=12)
ax.set_ylabel('True Label', fontsize=12)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("  [Saved] confusion_matrix.png")

# Classification Report
print("\n" + "="*55)
print("       CLASSIFICATION REPORT")
print("="*55)
print(classification_report(val_labels, pred_labels, target_names=CLASS_NAMES))


# ── STEP 9: Feature Map Visualization ────────────────────────
def visualize_feature_maps(model, image_array, class_names, image_idx=0):
    """
    Trích xuất và vẽ feature maps từ 3 lớp Conv2D đầu tiên.
    image_array: normalized (N, H, W, 3)
    """
    # Lấy các lớp Conv2D đầu
    conv_layer_names = [l.name for l in model.layers
                        if 'conv2d' in l.name and not l.name.endswith('b')][:3]
    conv_outputs = [model.get_layer(n).output for n in conv_layer_names]

    feat_model = Model(inputs=model.inputs, outputs=conv_outputs)

    # Chuẩn bị ảnh
    img = image_array[image_idx]
    img_tensor = np.expand_dims(img, axis=0)   # (1, 150, 150, 3)

    feature_maps = feat_model.predict(img_tensor, verbose=0)
    if not isinstance(feature_maps, list):
        feature_maps = [feature_maps]

    for fmap, name in zip(feature_maps, conv_layer_names):
        n_maps = min(fmap.shape[-1], 16)
        cols = 4
        rows = int(np.ceil(n_maps / cols))
        fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 2.8))
        fig.suptitle(f"Feature Maps — {name}  (shape {fmap.shape[1:]})",
                     fontsize=13, fontweight='bold')
        for i in range(rows * cols):
            ax = axes[i // cols][i % cols] if rows > 1 else axes[i % cols]
            if i < n_maps:
                fm = fmap[0, :, :, i]
                fm = (fm - fm.min()) / (fm.max() - fm.min() + 1e-8)
                ax.imshow(fm, cmap='viridis')
                ax.set_title(f"Filter {i+1}", fontsize=8)
            ax.axis('off')
        plt.tight_layout()
        save_name = f'feature_map_{name}.png'
        plt.savefig(save_name, dpi=120, bbox_inches='tight')
        plt.show()
        print(f"  [Saved] {save_name}")

    # Hiển thị ảnh gốc
    fig, ax = plt.subplots(1, 1, figsize=(4, 4))
    ax.imshow(image_array[image_idx])
    ax.set_title(f"Input: {class_names[val_labels[image_idx]]}", fontsize=12)
    ax.axis('off')
    plt.tight_layout()
    plt.savefig('feature_map_input.png', dpi=120)
    plt.show()


print("\n Visualizing Feature Maps...")
visualize_feature_maps(model, val_images, CLASS_NAMES, image_idx=0)


# ── STEP 10: Dự đoán ảnh mới ─────────────────────────────────
def predict_single_image(model, image_path, class_names, image_size=(150, 150)):
    """
    Dự đoán một ảnh từ đường dẫn file.
    Trả về (class_name, confidence_percent)
    """
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Không đọc được ảnh: {image_path}")
    img_rgb  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_res  = cv2.resize(img_rgb, image_size)
    img_norm = img_res.astype('float32') / 255.0
    img_inp  = np.expand_dims(img_norm, axis=0)   # (1, 150, 150, 3)

    probs      = model.predict(img_inp, verbose=0)[0]
    pred_idx   = np.argmax(probs)
    pred_class = class_names[pred_idx]
    confidence = probs[pred_idx] * 100.0

    # Hiển thị
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    axes[0].imshow(img_rgb)
    axes[0].set_title(f"Input Image", fontsize=12)
    axes[0].axis('off')

    colors = ['#E74C3C' if i == pred_idx else '#95A5A6' for i in range(len(class_names))]
    axes[1].barh(class_names, probs * 100, color=colors)
    axes[1].set_xlabel("Confidence (%)")
    axes[1].set_title(f"Kết quả: {pred_class}  ({confidence:.1f}%)", fontsize=12)
    axes[1].set_xlim([0, 100])
    plt.tight_layout()
    plt.show()

    return pred_class, confidence


def predict_batch_images(model, image_dir, class_names, image_size=(150, 150), max_show=15):
    """
    Dự đoán hàng loạt ảnh từ một thư mục.
    """
    exts = ('.jpg', '.jpeg', '.png', '.bmp', '.webp')
    files = [f for f in os.listdir(image_dir) if f.lower().endswith(exts)]
    if not files:
        print(f"Không có ảnh trong thư mục: {image_dir}")
        return

    images, filenames = [], []
    for f in files:
        img = cv2.imread(os.path.join(image_dir, f))
        if img is None:
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, image_size)
        images.append(img)
        filenames.append(f)

    images_arr  = np.array(images, dtype='float32') / 255.0
    raw_preds   = model.predict(images_arr, verbose=0)
    pred_labels = np.argmax(raw_preds, axis=1)

    # Print
    print(f"\n{'File':<30} {'Predicted':<15} {'Confidence':>10}")
    print("-" * 58)
    for fname, pred, prob in zip(filenames, pred_labels, raw_preds):
        print(f"{fname:<30} {class_names[pred]:<15} {prob[pred]*100:>9.1f}%")

    # Visualize
    n_show = min(len(images), max_show)
    cols   = 5
    rows   = int(np.ceil(n_show / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 3.2))
    fig.suptitle("Dự đoán hàng loạt ảnh mới", fontsize=14, fontweight='bold')
    for i in range(rows * cols):
        ax = axes[i // cols][i % cols] if rows > 1 else axes[i % cols]
        if i < n_show:
            ax.imshow(images[i])
            conf = raw_preds[i][pred_labels[i]] * 100
            ax.set_title(f"{class_names[pred_labels[i]]}\n{conf:.1f}%", fontsize=9)
        ax.axis('off')
    plt.tight_layout()
    plt.savefig('batch_predictions.png', dpi=130, bbox_inches='tight')
    plt.show()
    print("  [Saved] batch_predictions.png")

    return pred_labels, raw_preds


# Ví dụ sử dụng:
# pred_class, conf = predict_single_image(model, 'path/to/nail.jpg', CLASS_NAMES)
# print(f"Kết quả: {pred_class}  ({conf:.1f}%)")

# predict_batch_images(model, 'DataSet/Test', CLASS_NAMES)


# ── STEP 11: Lưu mô hình ─────────────────────────────────────
model.save(MODEL_SAVE_PATH)
print(f"\n Mô hình đã được lưu: {MODEL_SAVE_PATH}")

# Lưu thêm định dạng SavedModel
model.save('/content/drive/MyDrive/DataDeeplearning/Models/nail_cnn_model.keras')
print(" Mô hình đã được lưu: /content/drive/MyDrive/DataDeeplearning/Models/nail_cnn_model.keras")

print("\n" + "="*55)
print("   HOÀN THÀNH ĐỒ ÁN CNN NHẬN DIỆN MÓNG TAY!")
print("="*55)
```

```python
# ============================================================
#  predict.py — Dự đoán ảnh móng tay bằng model đã train
# ============================================================

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

# ── CONFIG ───────────────────────────────────────────────
MODEL_PATH = '/content/drive/MyDrive/DataDeeplearning/Models/nail_cnn_model.keras'   # chỉnh lại nếu cần
IMAGE_SIZE = (150, 150)

CLASS_NAMES = ['Long_Nail', 'Painted_Nail', 'Short_Nail']


# ── LOAD MODEL ───────────────────────────────────────────
def load_trained_model(model_path):
    if not os.path.exists(model_path):
        raise FileNotFoundError(f" Không tìm thấy model: {model_path}")
    print(" Đang load model...")
    model = keras.models.load_model(model_path)
    print(" Load model thành công!")
    return model


# ── PREPROCESS ───────────────────────────────────────────
def preprocess_image(image_path, image_size):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f" Không đọc được ảnh: {image_path}")

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, image_size)
    img = img.astype('float32') / 255.0

    return img


# ── PREDICT 1 IMAGE ──────────────────────────────────────
def predict_single(model, image_path):
    img = preprocess_image(image_path, IMAGE_SIZE)
    img_input = np.expand_dims(img, axis=0)

    probs = model.predict(img_input, verbose=0)[0]
    pred_idx = np.argmax(probs)

    pred_class = CLASS_NAMES[pred_idx]
    confidence = probs[pred_idx] * 100

    print(f"\n Ảnh: {image_path}")
    print(f" Dự đoán: {pred_class} ({confidence:.2f}%)")

    # Hiển thị ảnh
    plt.imshow(img)
    plt.title(f"{pred_class} ({confidence:.1f}%)")
    plt.axis('off')
    plt.show()

    return pred_class, confidence


# ── PREDICT FOLDER ───────────────────────────────────────
def predict_folder(model, folder_path):
    exts = ('.jpg', '.jpeg', '.png', '.bmp', '.webp')

    files = [f for f in os.listdir(folder_path) if f.lower().endswith(exts)]
    if not files:
        print(" Không có ảnh trong thư mục!")
        return

    print(f"\n Đang dự đoán {len(files)} ảnh...\n")

    for file in files:
        path = os.path.join(folder_path, file)
        try:
            predict_single(model, path)
        except Exception as e:
            print(f" Lỗi với file {file}: {e}")


# ── MAIN ─────────────────────────────────────────────────
if __name__ == "__main__":
    model = load_trained_model(MODEL_PATH)

    # ===== CHỌN CÁCH DÙNG =====

    # 1. Dự đoán 1 ảnh
    predict_single(model, '/content/drive/MyDrive/DataDeeplearning/DataSet/Train/Painted_Nail/H20.jpg')

    # 2. Dự đoán cả folder
    #predict_folder(model, '/content/drive/MyDrive/DataDeeplearning/DataSet/Train/Painted_Nail')
```
